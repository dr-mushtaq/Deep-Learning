**1D Vector in PyTorch**

```python
import torch
marks = torch.tensor([78, 85, 90, 88, 92])
print(marks)
print(marks.shape)
```

**2D Matrix in PyTorch**

```python
import torch
student_marks = torch.tensor([
    [78, 85, 90],
    [88, 92, 80],
    [75, 79, 95]
])
print(student_marks)
print(student_marks.shape)
```
